```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import os
```

```python
os.getcwd()
```

```python
covid_juli=pd.read_csv('C:\\Users\\Matthew\\Documents\\covid_juli.csv',parse_dates=['Tanggal'])
covid_juli.head()
```

```python
plt.plot(covid_juli['Tanggal'],covid_juli['Total kasus'],linewidth=2,label='Positif')
plt.plot(covid_juli['Tanggal'],covid_juli['Sembuh'],linewidth=2,color='green',label='Sembuh')
plt.plot(covid_juli['Tanggal'],covid_juli['Meninggal\r\r\nDunia'],linewidth=2,color='red',label='Meninggal')
plt.legend(loc='upper left')

plt.stackplot(covid_juli['Tanggal'],covid_juli['Total kasus'],alpha=0.6,color='#1f77b4')
plt.stackplot(covid_juli['Tanggal'],covid_juli['Sembuh'],alpha=0.6,color='green')
plt.stackplot(covid_juli['Tanggal'],covid_juli['Meninggal\r\r\nDunia'],alpha=0.6,color='red')
plt.xlabel('Tanggal')
plt.ylabel('Total Kasus')
plt.title('(Juli 2020)')
plt.suptitle('Total Kasus COVID-19 di Indonesia')
plt.xticks([737607., 737615.,737623.,737631.,737638.])
plt.show()
```

```python
plt.bar(covid_juli['Tanggal'],covid_juli['Kasus baru'],alpha=0.6,label='Positif')
plt.bar(covid_juli['Tanggal'],covid_juli['Sembuh\r\r\n(baru)'],alpha=0.6,color='green',label='Sembuh')
plt.bar(covid_juli['Tanggal'],covid_juli['Meninggal\r\r\n(baru)'],alpha=0.6,color='red',label='Meninggal')
plt.legend(loc='upper left',prop={'size': 8})

plt.xlabel('Tanggal')
plt.ylabel('Kasus Baru')
plt.title('(Juli 2020)')
plt.suptitle('Perkembangan Kasus Baru COVID-19')
plt.xticks([737607., 737615.,737623.,737631.,737638.])
plt.show()
```

```python
covid_prov=pd.read_csv('C:\\Users\\Matthew\\Documents\\heatmap.csv',parse_dates=['Tanggal'])

covid_prov['Hari']=covid_prov['Tanggal'].dt.day
covid_prov.head()
```

```python
heatmap_data = pd.pivot_table(covid_prov, values='Jumlah', 
                     index=['Provinsi'], 
                     columns='Hari')
plt.figure(figsize=(20,10))
sns.heatmap(heatmap_data,cmap="Blues",linewidths=0.5)
plt.xlabel('Tanggal')
plt.title('Kasus Positif COVID-19 Berdasarkan Provinsi')
plt.show()
```
